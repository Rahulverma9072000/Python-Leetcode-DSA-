```python
## Arithmetic operations
a = 12
b = 5
add_result = a + b
sub_result = a - b
mult_result = a * b
div_result = a / b
floor_div_result = a // b  # It will give the floor value of the division result 
modulus_result = a % b
exponent_result = a ** b
print(add_result , sub_result , mult_result , div_result , floor_div_result , modulus_result,exponent_result)
```

```python
# Comparison Operators 
## Equal to Operators 
a = 10 
b = 10
a == b
```

```python
str1 = "Rahul"
str2 = "rahul"
str1 == str2
```

```python
# Not Equals to 
str1 != str2
```

```python
# Greater than Operation 
num1 =  45 
num2 = 55
num1 > num2 
```

```python
#Less than Operation 
num1 < num2 
```

```python
## Logical Oeprators 
x = True 
y = True 
print(x and y)
print(x or y)
print(not x)
```

```python
x = False
y = False
print(x and y)
print(x or y)
print(not x)
```

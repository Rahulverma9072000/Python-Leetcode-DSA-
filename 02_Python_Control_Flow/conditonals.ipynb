```python
# if Statements 
age = 20
if age >= 18:
    print("Elegible to Vote Now")
```

```python
# else 
# The else statements executes a block of code if the condition in the if statement is false 

age = 16 
if age >= 18:
    print("Elegible to Vote Now")
else:
    print("Not Elegible for Now")
```

```python
# elif 
## THe elif Statements allows you to check multiple conditions . It stands for "else if" 

age = 20 
if age < 13:
    print("You are Child")
elif age < 18:
    print("You are teenager")
else:
    print("You are an Adult")
```

```python
# Nested COnditional Statements 
# Number Even , Odd , Negative 

num = int(input("Enter the Number"))
if num >= 0:
    print("The Number is Positive")
    if num %2 == 0:
        print("Number is Even")
    else:
        print("Number is Odd")
else:
    print("THe Number is Negative")
    if num %2 == 0:
        print("Number is Even")
    else:
        print("Number is Odd")
```
